```python
def generate_primes(l):
    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    primes = []
    n = 0
    while len(primes) < ((2**l)*2)-2:
        if is_prime(n):
            primes.append(n)
        n += 1
    return primes

```

```python
def remove_first_n(lst, n):
    # print("***********************N:", n)
    if len(lst) != 2:
        return lst[n:]
    else:
        return lst
```

```python
def last_level_node(lst, n):
    p = 2**n
    new_edge_list = remove_first_n(lst, p)
    return new_edge_list, lst[:p], lst[:p]
```

```python
def generate_edge(lst, n):
    p = 2**n
    new_edge_list = remove_first_n(lst, p)
    return new_edge_list, lst[:p]
```

```python
def Return_Single_Value(upper_edges, index):
    return upper_edges[index]
```

```python
def generate_node(upper_edges, lower_edges):
    nodes = []
    p = 0
    for x,y in zip(*[iter(lower_edges)]*2):
   
        i = Return_Single_Value(upper_edges,p)
        nodes.append(x+y+i)
        p += 1
    return nodes
    
```

```python
def get_root_node(edge_list):
    root_node=0
    for i in edge_list:
        root_node += i
    return root_node
```

```python
def sieve_of_eratosthenes(limit):
    primes = []
    is_prime = [True] * (limit + 1)
    for num in range(2, int(limit**0.5) + 1):
        if is_prime[num]:
            primes.append(num)
            for multiple in range(num * num, limit + 1, num):
                is_prime[multiple] = False
    return primes

def generate_primes(l):
    limit = ((2**l) * 2) - 2
    primes = sieve_of_eratosthenes(limit)
    return primes

def remove_first_n(lst, n):
    if len(lst) != 2:
        return lst[n:]
    else:
        return lst

def last_level_node(lst, n):
    p = 2**n
    new_edge_list = remove_first_n(lst, p)
    return new_edge_list, lst[:p], lst[:p]

def generate_edge(lst, n):
    p = 2**n
    new_edge_list = remove_first_n(lst, p)
    return new_edge_list, lst[:p]

def generate_node(upper_edges, lower_edges, p):
    nodes = []
    for x, y in zip(*[iter(lower_edges)] * 2):
        if p < len(upper_edges):
            i = upper_edges[p]
            nodes.append(x + y + i)
            p += 1
        else:
            break
    return nodes

def get_root_node(edge_list):
    return sum(edge_list)

all_nodes = []

def perfectBinaryTree(lev):
    all_nodes.clear()
    level = lev
    latest_lower_edge_list = []
    edges = generate_primes(level)

    if level < 0:
        print("Enter a positive integer!")
        return 0
    if level == 0:
        print("Root node: 2")
        all_nodes.append(2)
        return 0
    else:
        new_edges, latest_lower_edge_list, latest_node_list = last_level_node(edges, level)
        all_nodes.extend(latest_lower_edge_list)
        print("Level -", level, ':')
        p = level - 1
        while p >= 0:
            if p > 0:
                new_edges, latest_upper_edge_list = generate_edge(new_edges, p)
                nodes = generate_node(latest_upper_edge_list, latest_lower_edge_list, p)  # Pass p as an argument
                all_nodes.extend(nodes)
                p -= 1
                continue

            if p == 0:
                root_node = get_root_node(new_edges)
                all_nodes.append(root_node)
                break
    
    print("Number of initial nodes:", len(all_nodes))
    print("Number of unique nodes:", len(set(all_nodes)))

level_to_generate = 27
s = "For graph of level-"
print("\033[1m" + s + "\033[0m", level_to_generate)
perfectBinaryTree(level_to_generate)
```
